```python
average_transaction_amount = retail_df['TotalAmount'].mean()  # Changed df to retail_df and 'Revenue' to 'TotalAmount'
print(f"Average Transaction Amount: {average_transaction_amount:}")
```

```python
high_value_transactions = retail_df[retail_df['TotalAmount'] > 200]  # Use retail_df and 'TotalAmount'
print("\nTransactions with Total Amount greater than 200:")
print(high_value_transactions)
```

```python
customer_spending = retail_df.groupby('CustomerID')['TotalAmount'].sum()  # Changed df to retail_df and 'Revenue' to 'TotalAmount'
top_3_customers = customer_spending.nlargest(3)
print("\nTop 3 Customers who have spent the most:")
print(top_3_customers)
```
